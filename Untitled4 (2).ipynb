```python
# scatter plot
# Matplotlib, uninteractive
plt.scatter(data["Danceability"],data["Energy"], s=10)
plt.xlabel("Danceability")
plt.ylabel("Energy")
plt.title("Relationship between Danceability and Energy")
plt.show()
```

```python
# plotly, interactive
fig = px.scatter(data, x="Danceability", y="Energy", color=data["Key"], 
        title="Relationship between Danceability and Energy")
fig.show()
```

```python
# seaborn
sns.scatterplot(data=data, x="Danceability",y="Energy")
plt.xlabel("Danceability")
plt.ylabel("Energy")
plt.title("Relationship between Danceability and Energy")
plt.show()
```

```python
# hvplot, interactive
data.hvplot.scatter(x="Danceability", y="Energy", color="Key", marker="x",
                          title="Relationship between Danceability and Energy")
```

```python
# Question2
# correlation between loudness and speech
df2 = data.select_dtypes(include="number")
df2_corr = df2.corr().round(2)
df2_corr = df2.corr().apply(lamda nums: round(nums, 2))
# Correlation graph, using heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df2_corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
#pairplot
sns.pairplot(df2)
plt.show()
```

```python
#distribution of hits
df3 = data.groupby("Year")["Track"].count().reset_index()
df3
fig = sns.barplot(data=df3, x="Year", y="Track")
for i in fig.containers:
    fig.bar_label(i, label_type="edge")
plt.xlabel("Year")
plt.ylabel("Number of Hits")
plt.show()
```

```python
#popularity of Artists based on popularity
data.head()
most_pop_artiste = data.groupby('Artist')[['Popularity']].sum()
most_pop_artiste.sort_values(by='Popularity', ascending=False)
top_10 = most_pop_artiste.sort_values(by='Popularity', ascending=False)[0 : 10]
top_10
sns.set(rc={'figure.figsize': (12, 9)})
fig = sns.barplot(top_10, x='Artist', y="Popularity")
for i in fig.containers:
    fig.bar_label(i, label_type='edge')
plt.xticks(rotation=90, ha='center')
plt.ylabel("Popularity Score")
plt.title("Top 10 Artists by Popularity, 1990-1999")
# plt.savefig("most_pop.png")
plt.show()



```
